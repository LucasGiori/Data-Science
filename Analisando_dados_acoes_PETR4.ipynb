# Customizando Faixas de Eixos (Range)

```
import plotly.graph_objects as go

Close = go.Scatter(
                    x=df.Date, #Definindo Eixo X a coluna data do arquivo CSV
                    y=df.Close,#Definindo o Eixo Y com a coluna Close 
                   name="PETR High",
                   line = dict(color = '#17BECF'),
                   opacity=0.8)

data = [Close]
layout = dict(
    title="Série com Rangeslider e Botões",
    title_x=0.5,
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                          dict(count=1,
                               label='1 Mês',
                               step='month',
                               stepmode='backward'
                          ),
                          dict(
                              count=6,
                              label='6 Meses',
                               step='month',
                               stepmode='backward'
                          ),
                          dict(step='all')                          
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type='date'
    )
)
fig = dict(data=data, layout=layout)
py.iplot(fig)
```

# Trabalhando com Candlesticks

```
trace = go.Candlestick(x=df['Date'],
                       open=df['Open'],
                       high=df['High'],
                       low=df['Low'],
                       close=df['Close'])
data = [trace]
py.iplot(data,filename='simple_candlestick')
```

Candlesticks sem rangerslide

```
trace = go.Candlestick(x=df['Date'],
                       open=df['Open'],
                       high=df['High'],
                       low=df['Low'],
                       close=df['Close'])
layout = go.Layout(
    title = 'Preços em Candlesticks',
    title_x = 0.5,
    xaxis = dict(
        rangeslider = dict(
            visible = False
        )
    )
)
data = [trace]
fig = go.Figure(data=data,layout=layout)
py.iplot(fig)
```

## Customizando Candlesticks

```
trace = go.Candlestick(x=df['Date'],
                       open=df['Open'],
                       high=df['High'],
                       low=df['Low'],
                       close=df['Close'],
                       increasing=dict(line=dict(color= '#17BECF')),#Cor de fechamento positivo,
                       decreasing=dict(line=dict(color='#7F7F7F'))#Cor de fechamento Negativo
                       )
layout = go.Layout(
    title = 'Preços em Candlesticks',
    title_x = 0.5,
    xaxis = dict(
        rangeslider = dict(
            visible = False
        )
    )
)
data = [trace]
fig = go.Figure(data=data,layout=layout)
py.iplot(fig)
```

# Trabalhando Com Médias

```
df.head()
```

```
#Método mean - Média
#Média Simples de 3 Dias
df['MM_3'] = df.Close.rolling(window=3).mean()

#Média Simples de 9 dias
df['MM_9'] = df.Close.rolling(window=9).mean()

# Média Simples de 17 dias
df['MM_17'] = df.Close.rolling(window=17).mean()
```

```
df.head(20)
```

# Plotar as Médias e os Preços

```
close = go.Scatter(
    x=df.Date,
    y=df.Close,
    name = "PETR4 Close",
    line = dict(color = "#330000"),
    opacity = 0.8
)

MM_3 = go.Scatter(
    x=df.Date,
    y=df['MM_3'],
    name="Média Móvel 3 Períodos",
    line = dict(color = "#000080"),
    opacity = 0.8
)

MM_9 = go.Scatter(
    x=df.Date,
    y=df['MM_9'],
    name="Média Móvel 9 Períodos",
    line = dict(color = "#B2FF66"),
    opacity = 0.8
)

MM_17 = go.Scatter(
    x=df.Date,
    y=df['MM_17'],
    name="Média Móvel 17 Períodos",
    line = dict(color = "#B2FF66"),
    opacity = 0.8
)

data = [close,MM_3,MM_9,MM_17]
py.iplot(data)
```
