```
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```
pytrend.build_payload(kw_list=['Crise'])
```

```
# Interesse por Regi√£o
df = pytrend.interest_by_region()
```

```
df.head(10)
```

```
df.reset_index().plot(x='geoName', y='Crise', figsize=(120, 10), kind ='bar')
```

Principais Tend√™ncias de pesquisas di√°rias


```
df = pytrend.trending_searches(pn='brazil') #Se quiser do mundo deixar par√¢metro em branco, #united_states
df.head(100)
```

Sugest√£o de palavras chaves a partir de uma palavra

```
keywords = pytrend.suggestions(keyword='bolsonaro')
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')   # Excluindo coluna mid retornada
```

Consultas Relacionadas a uma pesquisa..

```
pytrend.build_payload (kw_list = ['Coronavirus'])
# Consultas relacionadas, retorna um dicion√°rio de quadros de dados 
related_queries = pytrend.related_queries()
related_queries.values()
```
